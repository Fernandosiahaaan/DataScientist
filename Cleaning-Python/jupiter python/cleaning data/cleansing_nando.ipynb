```python
data = data.dropna(axis = 1, how = "all")  #hapus kolom yang nilai nya nan semua
data
```

```python
data["Final Population"] = data["Final Population"]
for column in data.columns:
    # print(column)
    # data[column] = data[column].str.replace(",","")
    data[column] = data[column].replace(",","", regex = True) # menghapus nilai comma

data
```
